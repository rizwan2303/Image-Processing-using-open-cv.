# Types of Image thresholing(with Matplotlib)

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg', -1)
cv2.imshow('image', img)                  # Displaying the image using 'imshow' (RGB format)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Displaying the image using plt from matplotlib, it dispalys the image in BGR format.

plt.imshow(img)                         
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img)  
plt.xticks([]), plt.yticks([])       #axes are removed
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Displaying the image after converting in RGB format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)                          
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Performing various types of image Thresholding and displaying all of them using matplotlib.
```

```python
img = cv2.imread('gradient.png',0)
_, th1 = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY)
_, th2 = cv2.threshold(img, 200, 255, cv2.THRESH_BINARY_INV)
_, th3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
_, th4 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
_, th5 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, th1 ,th2 ,th3 ,th4, th5]

for i in range(6):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
    
plt.show()    

# For displaying using 'imshow'
#cv.imshow("th1", th1)
#cv.imshow("th2", th2)
#cv.imshow("th3", th3)
#cv.imshow("th4", th4)
#cv.imshow("th5", th5)

#cv.waitKey(0)
#cv.destroyAllWindows()
```
