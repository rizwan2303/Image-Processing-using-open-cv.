```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Reading the image
img = cv2.imread('opencv-logo.png')

# Coverting the image to RGB format as cv2 reads it in BGR format and matplotlin reads the image in RGB format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# defining a mask for average filtering operation
mask = np.ones((5, 5), np.float32)/25

# Average filtering operation on loaded image with mask
dst = cv2.filter2D(img, -1, mask)


# Blurring operation using pred defined blur function 
blur = cv2.blur(img, (5, 5))

# Gaussian Blurring operation using pred defined Gaussian blur function 
gblur = cv2.GaussianBlur(img, (5, 5), 0)

# Median filtering operation using pred defined median filtering function 
median = cv2.medianBlur(img, 5)

# Bilateral filtering operation using pred defined Bilateral filtering function 
bilateralFilter = cv2.bilateralFilter(img, 9, 75, 75)


# Defining a list called titles which will be used in for loop to give the title to images

titles = ['image', '2D Convolution', 'blur', 'GaussianBlur', 'median', 'bilateralFilter']


# Defining a list called images which will be used in for loop to acess the results of various filtering opereations done

images = [img, dst, blur, gblur, median, bilateralFilter]


for i in range(6):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
